```python
import requests
from bs4 import BeautifulSoup
```

```python
r1 = requests.get('http://172.18.0.2/login.php')
cookies = r1.cookies
print r1.text
```

```python
html = BeautifulSoup(r1.text, 'html.parser')
user_token = html.find("input", {"name":"user_token"})['value']
print user_token
```

```python
payload = {'username':'admin', 'password': 'password', 'Login': 'Login', 'user_token': user_token}
headers = r1.headers.copy()
headers['Content-Type'] = 'application/x-www-form-urlencoded'

r2 = requests.post('http://172.18.0.2/login.php', data=payload, cookies=cookies, headers=headers)
```
